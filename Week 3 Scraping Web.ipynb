```python
#Import Library
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
```

```python
alamat = "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)"
req = Request(alamat, headers={'User-Agent': 'Mozilla/5.0'})
```

```python
html = urlopen(req).read()
data = BeautifulSoup(html, 'html.parser')
table = data.find_all('table', {'class': 'wikitable sortable static-row-numbers plainrowheaders srn-white-background'})[0]
```

```python
rows = table.findAll('tr')
```

```python
hasil = []
for row in rows:
   cols = row.find_all('td')
   cols = [ele.text.strip() for ele in cols]
   hasil.append(cols)
```

```python
df_hasil = pd.DataFrame(hasil[2:])  # Mengambil baris mulai dari indeks 2 (baris ketiga)
df_hasil = df_hasil.reset_index(drop=True)  # Mengatur ulang indeks dengan drop=True
df_hasil
```

```python
# memberi judul kolom tabel
df_hasil.columns = ['Country/Territory', 'Region', 'IMF Forest', 'Year', 'World Bank Estimate', 'Year', 'UN Estimate', 'Year'] 
```
